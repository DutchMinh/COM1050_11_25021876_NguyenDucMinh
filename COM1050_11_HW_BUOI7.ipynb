```python
#w7a1
def binary_search(arr, target):
    l, r = 0, len(arr)
    while l <= r:
        mid = (l + r) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            r -= 1
        elif arr[mid] < target:
            l += 1
    return -1
num = list(map(int, input().split()))
target = int(input())
print(binary_search(num, target))
```

```python
#w7a2
def count_occurences(arr, x, count = 0):
    for i in arr:
        if i == x:
            count += 1
    return count
arr = list(map(int, input().split()))
x = int(input())
print(count_occurences(arr,x))
```

```python
#w7a3
def bubble_sort(arr):
    for i in range(len(arr)):
        swap = False
        for j in range(len(arr)-i-1):
            if arr[j] > arr [j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
        if not swap:
            return arr
def selection_sort(arr):
    for i in range(len(arr)):
        idx_min = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[i]:
                idx_min = j
        arr[i], arr[idx_min] = arr[idx_min], arr[i]
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))
```

```python
#w7a7
import ast
def longest_increase_sub(arr):
    arr = ast.literal_eval(arr)
    sub_len = [1] * len(arr)
    for i in range(len(arr)):
        for j in range(i):
            if arr[j] < arr[i]:
                sub_len[i] = max(sub_len[i], sub_len[j] + 1)
    return sub_len
nums = input()
print(longest_increase_sub(nums))
```

```python
#w7a8
import ast
interval = input()
queries = input()
interval = ast.literal_eval(interval)
queries = ast.literal_eval(queries)
def left_bound_query(arr, x):
    l, r = 0, len(arr)
    while l < r:
        mid = (l + r) // 2
        if arr[mid] < x:
            l = mid + 1
        else:
            r = mid
    return l 
def right_bound_query(arr, x):
    l, r = 0, len(arr)
    while l < r:
        mid = (l + r) // 2
        if arr[mid] <= x:
            l = mid + 1
        else:
            r = mid 
    return l - 1
def min_length_interval(arr):
    res = [10**18] * len(arr)
    for l,r in arr:
        left = left_bound_query(queries, l)
        right = right_bound_query(queries, r)
        if left <= right:
            length = right - left + 1
            for i in range(left, right + 1):
                res[i] = min(res[i], length)
    for i in range(len(arr)):
        if res[i] == 10**18:
            res[i] = -1
    return res
print(min_length_interval(interval))
```

```python
#w7a9
import ast
nums = input()
nums = ast.literal_eval(nums)
def smaller_count():
    sorted_lst = []
    res = [0] * len(nums)
    def lower_bound(arr, x):
        l, r = 0, len(arr)
        while l < r:
            mid = (l + r) // 2
            if arr[mid] <= x:
                l = mid + 1
            else:
                r = mid
        return r
    for i in range(len(nums)-1, -1, -1):
        position = lower_bound(sorted_lst, nums[i])
        if nums[i] not in sorted_lst:
            res[i] = position
        else:
            res[i] = res[i+1]
        sorted_lst.insert(position, nums[i])
    return res
print(smaller_count())
```

```python
#w7a10
import ast
nums = input()
m = int(input())
nums = ast.literal_eval(nums)
def upper_bound(arr, x):
    l, r = 0, len(arr)
    while l < r:
        mid = (l + r) // 2
        if arr[mid] <= x:
            l = mid + 1
        else:
            r = mid
        return r
prefix = [0] * len(nums)
prefix[0] = nums[0]
for i in range(1, len(nums)):
    prefix[i] = prefix[i-1] + nums[i]
def sub_sum_modul(arr):
    lst_mod = [0]
    res = []
    #a = prefix(arr)[r] % m
    #b = prefix(arr)[l-1] % m
    for r in range(len(arr)):
        a = prefix[r] % m
        idx = upper_bound(sorted(lst_mod),a)
        if idx < len(lst_mod):
            b = lst_mod[idx]
            res.append((a-b) % m)
        else:
            res.append(a % m)
        lst_mod.append(prefix[r] % m)
    return max(res)
print(sub_sum_modul(nums))
```
