```python
#w7a1
def binary_search(arr, target):
    l, r = 0, len(arr)
    while l <= r:
        mid = (l + r) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] > target:
            r -= 1
        elif arr[mid] < target:
            l += 1
    return -1
num = list(map(int, input().split()))
target = int(input())
print(binary_search(num, target))
```

```python
#w7a2
def count_occurences(arr, x, count = 0):
    for i in arr:
        if i == x:
            count += 1
    return count
arr = list(map(int, input().split()))
x = int(input())
print(count_occurences(arr,x))
```

```python
#w7a3
def bubble_sort(arr):
    for i in range(len(arr)):
        swap = False
        for j in range(len(arr)-i-1):
            if arr[j] > arr [j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
                swap = True
        if not swap:
            return arr
def selection_sort(arr):
    for i in range(len(arr)):
        idx_min = i
        for j in range(i+1, len(arr)):
            if arr[j] < arr[i]:
                idx_min = j
        arr[i], arr[idx_min] = arr[idx_min], arr[i]
    return arr
arr = list(map(int, input().split()))
print(bubble_sort(arr))
print(selection_sort(arr))
```

```python
#w7a4
def most_appear(arr):
    dct = {}
    for i in arr:
        if i not in dct:
            dct[i] = 1
        else:
            dct[i] += 1
    most_freq = []
    idx = []
    for k,v in dct.items():
        if v == max(dct.values()):
            most_freq.append(k)
            idx.append(arr.index(k))
    for i in most_freq:
        if arr.index(i) == min(idx):
            return (f"{i} xuat hien nhieu nhat, som nhat, {max(dct.values())} lan")
arr = list(map(int, input().split()))
print(most_appear(arr))
```
