```python
#w4a9
import math
while True:
    n = int(input())
    if n > 0:
        break
def perfect_square(m):
    for i in range(1,int(math.sqrt(m)) +1):
        if i ** 2 == m:
            return True
    return False
def digit_difference(x):
    s = str(x)
    digits = set()
    for y in s:
        if y in digits:
            return False
        digits.add(y)
    return True 
lst_perfect_square = []
for i in range (1, n+1):
    if perfect_square(i) and digit_difference(i):
        lst_perfect_square.append(i)
print(*lst_perfect_square, sep=" ")
```
