```python
#w4a9
import math
while True:
    n = int(input())
    if n > 0:
        break
def perfect_square(m):
    for i in range(1,int(math.sqrt(m)) +1):
        if i ** 2 == m:
            return True
    return False
def digit_difference(x):
    s = str(x)
    digits = set()
    for y in s:
        if y in digits:
            return False
        digits.add(y)
    return True 
lst_perfect_square = []
for i in range (1, n+1):
    if perfect_square(i) and digit_difference(i):
        lst_perfect_square.append(i)
print(*lst_perfect_square, sep=" ")
```

```python
#w4a32
password = str(input())
def char(n):
    if 97 <= ord(n) <= 122:
        return True
    return False
def num(n):
    if 48 <= ord(n) <= 57:
        return True
    return False
def capital(n):
    if 65 <= ord(n) <= 90:
        return True
    return False
def ky_tu_db(n):
    if not (num(n) or capital(n) or char(n)):
        return True
    return False
if len(password) <= 6:
    print('mat khau yeu')
else:
    def pass_strength():
        has_cap = False
        has_low = False
        has_num = False
        has_spe = False
        for i in password:
            if capital(i):
                has_cap = True
            elif char(i):
                has_low = True
            elif num(i):
                has_num = True
            elif ky_tu_db(i):
                has_spe = True
        return has_spe and has_cap and has_num and has_low
if pass_strength():
    print('mat khau manh')
else:
    print('mat khau yeu')
```
