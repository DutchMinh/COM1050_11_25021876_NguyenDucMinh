```python
#w5a3
def prime_num(n):
    if n == 1:
        return False
    else:
        for i in range(2,int((n ** 0.5)) + 1):
            if n % i == 0:
                return False
    return True
n = int(input())
print(prime_num(n))
```

```python
#w5a4
def perfect_num(n):
    lst_factor = []
    for i in range(1,n):
        if n % i == 0:
            lst_factor.append(i)
            if sum(lst_factor) == n:
                return True
    return False
n = int(input())
print(perfect_num(n))
```

```python
#w5a6
def factorial(n, s = 1):
    if n == 0:
        s = s
    else:
        for i in range(1,n + 1):
            s *= i
    return s
n = int(input())
print(factorial(n))
```

```python
#w5a9
def sum_digit(n):
    lst_digit = list(str(n))
    lst_digit = [int(n) for n in lst_digit]
    return sum(lst_digit)
n = int(input())
print(sum_digit(n))
```

```python
#w5a10
def dang_cau(a,b):
    mapAtoB = {}
    mapBtoA = {}
    for m,n in zip(a,b):
        if m in mapAtoB and mapAtoB[m] != n:
            return False
        else:
            mapAtoB[m] = n
        if n in mapBtoA and mapBtoA[n] != m:
            return False
        else:
            mapBtoA[n] = m
    return True
a, b = map(str, input().split())
print(dang_cau(a,b))
```
