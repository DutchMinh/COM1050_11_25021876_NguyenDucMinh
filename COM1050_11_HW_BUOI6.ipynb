```python
#w6a5
lst = list(map(int, input().split()))
diction = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}
for i in lst:
    if i > 0:
        diction['positives'] += 1
    elif i < 0:
        diction['negatives'] += 1
    else:
        diction['zeros'] += 1
print(diction)
```

```python
s1 = input()
s2 = input()
diction = {}
def pairs_define(s):
    s = s.strip('{}')
    if not s:
        return []
    pairs = s.split(',')
    lst_pairs = []
    for i in pairs:
        k, v = i.split(':')
        lst_pairs.append(f"{k}:{v}")
    return lst_pairs
lst_pairs = pairs_define(s1) + pairs_define(s2)
for i in lst_pairs:
    key, value = i.split(':')
    diction.setdefault(key, []).append(int(value))
for i in diction:
    total = sum(diction[i])
    diction[i] = total
dict_sorted = dict(sorted(diction.items()))
print(dict_sorted)
```
