```python
#w6a9
s1 = input()
s2 = input()
diction = {}
def pairs_define(s):
    s = s.strip('{}')
    if not s:
        return []
    pairs = s.split(', ')
    lst_pairs = []
    for i in pairs:
        k, v = i.split(':')
        k = k.strip("'")
        lst_pairs.append(f"{k}:{v}")
    return lst_pairs
lst_pairs = pairs_define(s1) + pairs_define(s2)
for i in lst_pairs:
    key, value = i.split(':')
    diction.setdefault(key, []).append(int(value))
for i in diction:
    total = sum(diction[i])
    diction[i] = total
print(dict(sorted(diction.items())))
```

```python
#w6a10
num = sorted(list(map(int, input().split())))
target = int(input())
lst_sum = []
check = set()
for a in num:
    b = target - a
    if a == b and num.count(a) < 2:
        continue
    if (min(a, b), max(a, b)) not in check and b in num:
        lst_sum.append((min(a, b), max(a, b)))
        check.add((min(a, b), max(a, b)))
    else:
        lst_sum = []
print(lst_sum)
```

```python
#w6a11
num = tuple(map(int, input().split()))
lst_even = []
lst_odd = []
for i in num:
    if i % 2 == 0:
        lst_even.append(i)
    else:
        lst_odd.append(i)
print(tuple(lst_even), tuple(lst_odd), sep = '\n')
```

```python
#w6a12
num = list(map(int, input().split()))
check = set()
freq = {}
for i in num:
    if i in check:
        continue
    count = num.count(i)
    freq[i] = count
    check.add(i)
max_count = max(freq.values())
most_frequent = [k for k, v in freq.items() if v == max_count]
print(min(most_frequent))
```
