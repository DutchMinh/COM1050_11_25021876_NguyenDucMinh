```python
#w6a1
lst1 = list(map(int, input().split()))
lst2 = []
for i in lst1:
    if i in lst2:
        continue
    else:
        lst2.append(i)
print(lst2)
```

```python
#w6a2
lst_num = list(map(int, input().split()))
total = 0
lst_sum = []
for i in lst_num:
    total += i
    lst_sum.append(total)
print(lst_sum)
```

```python
#w6a3
tpl = tuple(map(int, input().split()))
k = int(input())
lst_num = list(tpl)
lst1 = lst_num[:k]
lst2 = lst_num[k:]
lst2.extend(lst1)
print(tuple(lst2))
```

```python
#w6a4
s = input()
pairs = s.split()
dict = {}
for i in pairs:
    key, value = i.split(':')
    if key not in dict:
        dict[key] = [value]
    else:
        dict[key].append(value)
print(dict)
```

```python
#w6a5
lst = list(map(int, input().split()))
dict = {
    'positives': 0,
    'negatives': 0,
    'zeros': 0
}
for i in lst:
    if i > 0:
        dict['positives'] += 1
    elif i < 0:
        dict['negatives'] += 1
    else:
        dict['zeros'] += 1
print(dict)
```
